```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
#Decision Tree Model, training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
dtree.fit(X_train,y_train)
```

## Prediction and Evaluation

```python
predictions = dtree.predict(X_test)
```

```python
print("Predictions array: ",predictions)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn import tree
plt.figure(figsize=(20,25))
tree.plot_tree(dtree,feature_names=X.columns,class_names=['Class-1', 'Class-0'],rounded=True, # Rounded node edges
          filled=True, # Adds color according to class
          proportion=True
        )
plt.show()
```

```python
user_input = [float(x) for x in input("Enter 30 feature values separated by commas: ").split(",")]
user_input_df = pd.DataFrame([user_input], columns=X.columns)  # Use same feature names
result = dtree.predict(user_input_df)

print("Prediction:", "Fraud" if result[0] == 1 else "Not Fraud")
```
